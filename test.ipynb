```python
import numpy as np
import os,sys
from PIL import Image
import six
import imagehash
import pandas as pd
import operator
```

```python
cats_dict={}
rootDir = 'F:\Cats and Dogs\cats'
j=0
for dirName, subdirList, fileList in os.walk(rootDir):
    #print('Found directory: %s' % dirName)
    for fname in fileList:
        #print('\t%s' % fname)
        cats_dict[j]=fname
        j=j+1
```

```python
dogs_dict={}
rootDir = 'F:\Cats and Dogs\dogs'
i=0
for dirName, subdirList, fileList in os.walk(rootDir):
    #print('Found directory: %s' % dirName)
    for fname in fileList:
        #print('\t%s' % fname)
        dogs_dict[i]=fname
        i=i+1
```

```python
list(dogs_dict.items())[11499][1]
```

```python
data_cats=[]
for i in range(10000):
    f='F:\Cats and Dogs\cats\\'+list(cats_dict.items())[i][1]
    data_cats.append(f)
```

```python
data_dogs=[]
for i in range(10000):
    f='F:\Cats and Dogs\dogs\\'+list(dogs_dict.items())[i][1]
    data_dogs.append(f)
```

```python
dogs=pd.DataFrame(data_dogs)
dogs['NC']='1'
dogs
```

```python
cats=pd.DataFrame(data_cats)
cats['NC']='0'
cats
```

```python
df=pd.concat([cats,dogs])
df.columns=["filename",'Category']
df
```
